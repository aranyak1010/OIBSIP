# DATA LOADING

```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
youtube = pd.read_csv('USvideos.csv')
```

# DATA DESCRIPTION

```python
print("Dimensions of the dataset: ", youtube.shape)
print("Size of the dataset: ",youtube.size)
print("Columns in the dataset: ",youtube.keys())
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset: \n")
print(youtube.head())
```

# DATA INTEGRITY

```python
# Check data types of each column
print("\nData types of each column:")
print(youtube.info())

# Get basic statistics for numerical columns
print("\nBasic statistics for numerical columns:")
print(youtube.describe())

# Check unique values in categorical columns
categorical_columns = ['video_id', 'trending_date', 'title', 'channel_title', 'category_id', 'publish_time', 'tags', 'thumbnail_link', 'description']
unique_values = {col: youtube[col].nunique() for col in categorical_columns}
print("\nUnique values in categorical columns:")
print(unique_values)
```

# HANDLING MISSING VALUES

```python
# Fill missing descriptions with a placeholder
youtube['description'].fillna('No description available', inplace=True)

# Verify if there are any missing values left
missing_data_check = youtube.isnull().sum()
print("\nMissing data check after handling missing values:")
print(missing_data_check)
```

# DUPLICATES REMOVAL

```python
# Check for duplicate records based on 'video_id' and 'trending_date'
duplicate_rows = youtube.duplicated(subset=['video_id', 'trending_date'])

# Count the number of duplicates
num_duplicates = duplicate_rows.sum()
print(f"\nNumber of duplicate records: {num_duplicates}")

# Remove duplicate records
data_cleaned = youtube.drop_duplicates(subset=['video_id', 'trending_date'])
print(f"\nShape of the dataset after removing duplicates: {data_cleaned.shape}")
```

# STANDARDIZATION

```python
# Convert 'trending_date' and 'publish_time' to proper datetime formats
data_cleaned['trending_date'] = pd.to_datetime(data_cleaned['trending_date'], format='%y.%d.%m')
data_cleaned['publish_time'] = pd.to_datetime(data_cleaned['publish_time'])

# Verify the changes
print("\nData types after standardization:")
print(data_cleaned.dtypes)
```

# DETECTING AND HANDLING OUTLIERS

```python
# Define a function to detect outliers using IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Identify outliers in numerical columns
numerical_columns = ['views', 'likes', 'dislikes', 'comment_count']
outliers_dict = {col: detect_outliers(data_cleaned, col) for col in numerical_columns}

print("\nNumber of outliers in each numerical column:")
for col, outliers in outliers_dict.items():
    print(f"{col}: {len(outliers)}")
```

```python
# Optionally, handle outliers by capping them
# Here, we will cap the outliers
for col in numerical_columns:
    Q1 = data_cleaned[col].quantile(0.25)
    Q3 = data_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data_cleaned[col] = data_cleaned[col].clip(lower_bound, upper_bound)

# Verify the changes
print("\nBasic statistics after handling outliers:")
print(data_cleaned.describe())
```
